```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
 

```

```python
#Seaborn also offers built-in data sets:
iris = sns.load_dataset("iris")
print(iris.info())
```

```python
#Set the matplotlib parameters
sns.set_style("ticks",  {"xtick.major.size":8,
"ytick.major.size":8})
```

```python
 #Plot data and regression model fits
 sns.lmplot(x="sepal_width", 
 y="sepal_length", 
 hue="species", 
 data=iris)
```

```python
#Scatterplot with one categorial variable
sns.stripplot(x="species", y="petal_length", data=iris) 
```

```python
#plot data and linear regression model fit
sns.regplot(x="sepal_width", 
 y="sepal_length", 
 data=iris)
```

```python
df=sns.load_dataset("tips")
print(df.info())
```

```python
#Show point estimates and confidence intervals with scatterplot glyphs 
sns.barplot(x="sex",y="tip",data=df)
```

```python
#A violin plot is a hybrid of a box plot and a kernel density plot, which shows peaks in the data.
sns.violinplot(x="time",y="size",data=df)
```

```python
#A boxplot is a standardized way of displaying the distribution of data based on a five number summary 
sns.boxplot(x="smoker",y="size",hue="time",data=df)
```

```python
#categorical scatterplot with non-overlapping points
sns.swarmplot(x="smoker",y="tip",data=df)
```

```python
#A histogram is a graph showing frequency distributions.
#It is a graph showing the number of observations within each given interval.
x = np.random.normal(170, 10, 250)
plt.hist(x)
plt.show()
```

```python
#Subplot grid for plotting conditional relationships.
g = sns.FacetGrid(df, 
 col="smoker", 
 row="sex")
g = g.map(plt.hist,"time")
```

```python
#Subplot grid for plotting pairwise relationships.
h = sns.PairGrid(iris) 
h = h.map(plt.scatter) 
```
