```python
"""Seaborn is a library that uses Matplotlib underneath to plot graphs. It will be used to visualize random distributionsSeaborn works by capturing entire dataframes or arrays containing all your data and performing all the internal functions necessary for semantic mapping and statistical aggregation to convert data into informative plots.."""
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
 titanic = sns.load_dataset("titanic")
print(titanic.info())
```

```python
 sns.set_style("ticks", {"xtick.major.size":8,"ytick.major.size":8})
```

```python
 g = sns.FacetGrid(titanic,col="survived",  row="sex")
```

```python
g = g.map(plt.hist,"age")
```

```python
sns.barplot(x="sex", y="survived", hue="class", data=titanic)
```

```python
 sns.pointplot(x="class", y="survived", 
 hue="sex", 
 data=titanic, 
 palette={"male":"g",
 "female":"m"},
 markers=["^","o"],
 linestyles=["-","--"])
```

```python
sns.boxplot(x="alive", 
 y="age",
 hue="adult_male",
 data=titanic)
```

```python
 sns.violinplot(x="age", 
 y="sex",
 hue="survived",
 data=titanic)
```
