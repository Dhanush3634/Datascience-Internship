```python
import pandas as pd
#here we imported pandas as pd
mydataset={
    'phone':["apple","nokia","Moto"],
    'Ram':["8","6","12"]
}
#here we initialised the dataframe by giving values to two attributes (phone,Ram)
print(pd.DataFrame(mydataset))
```

```python
import pandas as pd 
d=["dhanush","cherry","almond"]
myvar=pd.Series(d,index=["x","y","z"])
#here Series is used to create  column table and data type.
#here index is used to give labels to datavalues.
print(myvar)
#here we get the output of the datavalue whose index is "y".
print(myvar["y"])
#here we used key/value object like dictionaries.
Oscar={"movie-1":"avatar","movie-2":"joker","movie-3":"life of pie"}
print(pd.Series(Oscar))
```

```python
import pandas as pd
data={
    'bike':["cbr-250r","ducati","harley street "],
    'cc':[279,999,750]
}
#Here loc is used to get the specified row according to its index
print(pd.DataFrame(data).loc[1])
```

```python
import pandas as pd
data={
    'season':["summer","winter","rainy"],
    'weather':["hot","cold","rain"]
}
#with index argument,we can name our own indexes.
df=pd.DataFrame(data,index=["h","c","r"])
print(df)
#we can use named index with loc attribute , to return specified rows.
print(df.loc["h"])
```

```python
import pandas as pd
df=pd.read_csv('data.csv')
#fillna() method allows us to replace empty cells with a value.
df.fillna(130,inplace=True)
print(df.to_string())


```

```python
import pandas as pd
df = pd.read_csv('data.csv')
#This operation inserts 130 in empty cells in the "Calories" column (row 18 and 28)
df["Calories"].fillna(130, inplace = True)
print(df.to_string())
```

```python
import pandas as pd

df = pd.read_csv('data.csv')
#
x = df["Calories"].mean()
df["Calories"].fillna(x, inplace = True)
print("average value of column calories in empty cells ")
print(df.to_string())
y = df["Duration"].median()
df["Duration"].fillna(y, inplace = True)
print("print meadian value of column Duration in empty cells")
print(df.to_string())
z = df["Maxpulse"].mode()[0]
df["Maxpulse"].fillna(z, inplace = True)
print("mode value of column Maxpulse in empty cells ")      
print(df.to_string())







```

```python
import pandas as pd

df = pd.read_csv('data.csv')
#to discover duplicates, we can use the duplicated() method.

print(df.duplicated())
#To remove duplicates, use the drop_duplicates() method.
df.drop_duplicates(inplace = True)
print(df.to_string())
```
